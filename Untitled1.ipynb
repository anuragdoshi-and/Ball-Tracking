```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2 as cv
import os
import numpy as np
import re
```

```python
frames = os.listdir('frames/')
```

```python
frames.sort(key = lambda f: int(re.sub('frame#||.png', '',f)))
```

```python
images =[]
img = cv.imread('frame#0.png')

for i in frames:
    img = cv.imread('frames/'+ i)
    img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    img = cv.GaussianBlur(img,(25,25),0)
    images.append(img)

    
images = np.array(images)
```

```python
nonzero =[]
for i in range(len(images)-1):
    mask = cv.absdiff(images[i], images[i+1])
    _, mask = cv.threshold(mask, 50, 255, cv.THRESH_BINARY)
    num = np.count_nonzero(mask.ravel())
    nonzero.append(num)
   # print('test')
    
    
x = np.arange(0,len(images)-1)    
y = nonzero

plt.figure(figsize=(20,4))
plt.scatter(x,y)
```

```python
threshold = 15 * 10e3
for i in range(len(images)-1):
    if(nonzero[i]>threshold):
        scence_change_idx = i
        break
    
frames = frames[:(scence_change_idx+1)]
        
```

```python
img = cv.imread('frames/'+ frames[10])
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
gray = cv.GaussianBlur(gray, (25,25), 0)

plt.figure(figsize=(5,10))
plt.imshow(gray, cmap='gray')
```

```python
_, mask = cv.threshold(gray, 200, 255, cv.THRESH_BINARY)
plt.figure(figsize=(5,5))
plt.imshow(mask,cmap='gray')
```

```python
contours, images = cv.findContours(mask, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
img_copy = np.copy(gray)
img_copy = cv.drawContours(img_copy, contours, -1, (0,255,0), 3)
plt.imshow(img_copy, cmap= 'gray')
```

```python
num = 20 
cnt = 0

for i in range(len(contours)):
    x,y,w,h = cv.boundingRect(contours[i])
    
    numer = min([w,h])
    denom = max([w,h])
    ratio = numer/denom
    
    if(x>=num and y>= num):
        xmin, ymin = x-num, y-num
        xmax, ymax = x+w+num, y+h+num
        
    else:
        xmin, ymin = x , y
        xmax, ymax = x+w, y+h
    
    if(ratio>=0.5 and((w<=10) and (h<=10))):
        print(cnt, x, y, w, h, ratio)
        cv.imwrite("patch" + str(cnt) + ".png", img[ymin:ymax, xmin:xmax])
        cnt = cnt+1
```
